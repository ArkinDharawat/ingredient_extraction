```python
import pandas as pd
data = pd.read_csv("drive/MyDrive/682_BERT/combined11_23.csv",encoding="latin1" )
```

```python
data.head(30)
```

```python
data =data.fillna(method ="ffill")
```

```python
data.head(30)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data["Sentence"] = LabelEncoder().fit_transform(data["Sentence"] )
```

```python
data.head(80)
```

```python
data.rename(columns={"Sentence":"sentence_id","Word":"words","Tag":"labels"}, inplace =True)
```

```python
data["labels"] = data["labels"].str.upper() #might not be necessary 


X= data[["sentence_id","words"]]
Y =data["labels"]

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2, shuffle=False)

#building up train data and test data
train_data = pd.DataFrame({"sentence_id":x_train["sentence_id"],"words":x_train["words"],"labels":y_train})
test_data = pd.DataFrame({"sentence_id":x_test["sentence_id"],"words":x_test["words"],"labels":y_test})
```

```python
train_data
```

# Model Training

```python
from simpletransformers.ner import NERModel,NERArgs
```

```python
label = data["labels"].unique().tolist()
label
```

```python
# params to tune later?

args = NERArgs()
args.num_train_epochs = 3
args.learning_rate = 5e-5
args.overwrite_output_dir =True
args.train_batch_size = 32
args.eval_batch_size = 32
args.wandb_project = "BERT"
```

```python
model = NERModel('bert', 'bert-base-cased',labels=label,args =args)
```

```python
model.train_model(train_data,eval_data = test_data,acc=accuracy_score)
```

```python
result, model_outputs, preds_list = model.eval_model(test_data, output_dir = "drive/MyDrive/682_BERT/eval_output")
```

```python
ground_truth = []
sentences = []
sentence_index = 0


for test_idx in test_data.sentence_id.unique():
  ground_truth.append([])
  sentence = ""
  for row in test_data[test_data.sentence_id == test_idx].iterrows():
    token = row[1][1]
    ent = row[1][2]
    ground_truth[sentence_index].append(dict({token: ent}))
    sentence = sentence + token + " "
  sentence = sentence[:-1]
  sentences.append(sentence)
  sentence_index +=1
```

```python
print(ground_truth[0])
print(sentences[0])
```

```python
predictions, model_output = model.predict(sentences)
print(len(predictions))
print(len(ground_truth))
```

```python
print(len(ground_truth))
print(len(predictions))
```

```python
gt = ground_truth
bad = 0
for indx in range(len(gt)):
  # print(len(gt[indx]))
  # print(len(predictions[indx]))
  fp = 0
  fn = 0
  tp = 0
  tn = 0
  # gt_list = []
  # pred_list = []
  gt_inc = []
  pred_inc = []
  for dict_idx in range(len(predictions[indx])):
    # print(gt[indx][dict_idx])
    # print(predictions[indx][dict_idx])
    ent1 = list(gt[indx][dict_idx].keys())[0] 
    ent2 = list(predictions[indx][dict_idx].keys())[0]
    if gt[indx][dict_idx][ent1] != predictions[indx][dict_idx][ent2]:
      if gt[indx][dict_idx][ent1] =='O':
        fp +=1
        pred_inc.append(dict({ent2: predictions[indx][dict_idx][ent2]}))
        gt_inc.append(dict({ent1: gt[indx][dict_idx][ent1]}))
      else:
        fn +=1
        pred_inc.append(dict({ent2: predictions[indx][dict_idx][ent2]}))
        gt_inc.append(dict({ent1: gt[indx][dict_idx][ent1]}))
    if gt[indx][dict_idx][ent1] == predictions[indx][dict_idx][ent2]:
      if gt[indx][dict_idx][ent1] =='O':
        tn +=1
      else:
        tp +=1

  if (tp+fp == 0 or fn+tp == 0):
    precision = 0
    recall = 0
    f1 = 0
  else:
    precision = tp / (tp +fp)
    recall = tp / (fn + tp)
    if precision + recall == 0:
      f1 = 0
    else:
      f1 = (2*precision*recall)/(precision+recall)
  if f1 < 0.85:
    bad+=1
    print("Sentence:")
    print(sentences[indx])
    print("Ground Truth:")
    print(gt_inc)
    print("\nPrediction")
    print(pred_inc)
    print("---------------------------------------------------------------------------------------------------------------------------\n")

print(bad)
```

```python
# Running BERT on just curated and foodcom annotated (combined)

# Imports
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

#Load data
data = pd.read_csv("drive/MyDrive/682_BERT/combined_fbc_fca.csv",encoding="latin1" )

#Data preprocessing
data =data.fillna(method ="ffill")
data["Sentence"] = LabelEncoder().fit_transform(data["Sentence"] )
data.rename(columns={"Sentence":"sentence_id","Word":"words","Tag":"labels"}, inplace =True)

data["labels"] = data["labels"].str.upper() #might not be necessary 


X= data[["sentence_id","words"]]
Y =data["labels"]

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2)

#building up train data and test data
train_data = pd.DataFrame({"sentence_id":x_train["sentence_id"],"words":x_train["words"],"labels":y_train})
test_data = pd.DataFrame({"sentence_id":x_test["sentence_id"],"words":x_test["words"],"labels":y_test})

# Model Training
from simpletransformers.ner import NERModel,NERArgs

label = data["labels"].unique().tolist()

# params to tune later?
args = NERArgs()
args.num_train_epochs = 10
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 32
args.eval_batch_size = 32

model = NERModel('bert', 'bert-base-cased',labels=label,args =args)
```

```python
model.train_model(train_data,eval_data = test_data,acc=accuracy_score)
```

```python
result, model_outputs, preds_list = model.eval_model(test_data)
```
